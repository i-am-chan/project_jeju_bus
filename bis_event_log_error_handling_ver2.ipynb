```python
import pandas as pd
```

```python
prefix_path = 'D:/jeju_bus_data_no_leakage/bis/'
file_name = 'BIS_190530.csv'

path = prefix_path + file_name
```

```python
df = pd.read_csv(path, 
                 usecols = ['EVENT_DATE', 'EVENT_CD', 'VH_ID', 'ROUTE_ID', 
                            'BUS_RUN_CNT', 'ROUTE_RUN_CNT', 'RUN_DIST'])
```

```python
df = df[(df['EVENT_CD'] == 18) | (df['EVENT_CD'] == 19)]
```
