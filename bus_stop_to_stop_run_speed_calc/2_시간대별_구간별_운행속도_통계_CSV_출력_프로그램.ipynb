```python
import pandas as pd

prefix_path = ''
infix_path = 'BIS_' + '190531'
postfix_path = '_RUN_SPED_ANALYSIS.csv'
```

```python
path = prefix_path + infix_path + postfix_path

df = pd.read_csv(path)
```

```python
# default: 30분 간격 .... 1시, 1시 30분, 2시, 2시 30분, ...
```

```python
df.loc[:, 'EVENT_DATE'] = df['EVENT_DATE'].apply(lambda datetime: pd.to_datetime(datetime))
```

```python
df['EVENT_DATE'] = df['EVENT_DATE'].dt.hour + df['EVENT_DATE'].dt.minute//30/2
```

```python
grouped = df.groupby(by=['EVENT_DATE', 'STATION_ORD', 'STATION_ORD2'])
```

```python
grouped[['RUN_SPED']].mean()
```

개선 사항-> station_ord를 station_id 혹은 station_name으로 치환해야 위 코드가 의미있는 결과를 도출함.
