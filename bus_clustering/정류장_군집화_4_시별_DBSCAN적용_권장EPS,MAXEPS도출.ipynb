# DBSCAN-optimized eps 발견

## metric, minpts 상수

```python
metric = 'euclidean'
minPts = 3
```

### dataframe 읽어들이기

```python
path = r'D:\jeju_bus_data_no_leakage\station'.replace('\\', '/')

df = pd.read_csv(path + '/station_final.csv', encoding = 'ansi')

df['city'] = df['STATION_ADDR']
df.loc[:, 'city'] = df.loc[:, 'STATION_ADDR'].apply(lambda string: string.split(' ')[0])

seogwipo_city_station_df = df.query('city == "서귀포시"')
jeju_city_station_df = df.query('city == "제주시"')
```

### Seogwipo

```python
eps_set = [0.00055, 0.00065, 0.00075, 0.00085, 0.00095, 0.00105, 0.00115, 0.00125, \
           0.00135, 0.00145, 0.00155, 0.00165, 0.00175, 0.00185, 0.00195, 0.00205, 0.00215, 0.00225]
maxEps_set = [0.0022, 0.0023, 0.0024, 0.0025]
```

```python
for eps in eps_set:
    seogwipo = seogwipo_city_station_df.loc[:, ['LOCAL_X', 'LOCAL_Y']] # copy
    
    dbscan_lv1 = DBSCAN(eps = eps, min_samples = minPts + 1, metric = metric)
    dbscan_lv2 = DBSCAN(eps = eps, min_samples = minPts, metric = metric)
    
    dbscan_lv1_labels = dbscan_lv1.fit_predict(seogwipo)
    seogwipo['target'] = dbscan_lv1_labels 
    
    print(eps , '...' , 'lv1')
    print(seogwipo.groupby(by='target').count())
    
    seogwipo2 = seogwipo.query('target == -1')
    dbscan_lv2_labels = dbscan_lv2.fit_predict(seogwipo2)
    seogwipo2['target'] = dbscan_lv2_labels 
    
    print(eps , '...' , 'lv2')
    print(seogwipo2.groupby(by='target').count())
```
