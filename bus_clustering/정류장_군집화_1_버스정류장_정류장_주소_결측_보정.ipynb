# 1. 현재 사용 중인 버스정류장만 추출

```python
import pandas as pd

path = r'D:\jeju_bus_data_no_leakage\station'.replace('\\', '/')
file_name = path + '/station_제주연구원_20190531.csv'

df = pd.read_csv(file_name, 
                 usecols = ['STATION_ID', 'STATION_NM', 'CENTER_YN', 'LOCAL_X', 'LOCAL_Y', 'USE_YN', 'STATION_ADDR'],
                 encoding = 'ansi')

df = df.query("USE_YN == 'Y'")
```

```python
df.head(5) # 테스트
```

# 2. 버스정류장 주소(STATION_ADDR)가 
# 결측된 행만 추출 → csv로 저장

```python
addr_correct_df = df[df['STATION_ADDR'].notnull()] # 정류장주소가 결측되지 않은 행 추출
```

```python
addr_missing_df = df[df['STATION_ADDR'].isnull()][df.columns[:-1]]
```

```python
addr_missing_df.head(5) # 테스트
```

```python
addr_missing_df.to_csv(path + '/' + 'station_addr_missing_df.csv', index=False, encoding='ansi')
```

# 4. station_addr_tidying.shp.csv EDA

```python
addr_tidying_df = pd.read_csv(path + '/' + 'station_addr_tidying_df.shp.csv', encoding = 'ansi')
```

```python
del addr_tidying_df['No']
```

```python
addr_tidying_df[addr_tidying_df['도로명주소'].isnull()] # 테스트
```

```python
addr_tidying_df[addr_tidying_df['지번주소'].isnull()] # 테스트
```

## result.shp.csv 확인 결과 '지번주소' 변환 실패 건수가 
## 더 적으므로 '지번주소'를 'STATION_ADDR'로 사용.
## (단, 지번주소가 NaN인 행은 도로명주소 필드를 사용.)

```python
jibeon_juso_missing_df = addr_tidying_df[addr_tidying_df['지번주소'].isnull()]
```

```python
addr_tidying_df.loc[jibeon_juso_missing_df.index, '지번주소'] =\
    jibeon_juso_missing_df.loc[:, '도로명주소']
```

```python
addr_tidying_df.query('STATION_ID == 405002394') # 테스트 (지번주소가 tidying되었나)
```

```python
addr_tidying_df = addr_tidying_df.rename(columns = {'지번주소': 'STATION_ADDR'})
```

```python
addr_tidying_df.loc[:, 'STATION_ADDR'] =\
    addr_tidying_df['STATION_ADDR'].apply(lambda obj: ' '.join(str(obj).split(' ')[1:]))
```

```python
del addr_tidying_df['도로명주소']
```

```python
addr_tidying_df.head() # 테스트
```

# 5. addr_tidying_df와 addr_correct_df 합집합(concat)

```python
result = pd.concat([addr_correct_df, addr_tidying_df], ignore_index=True)
result
```

```python
result.to_csv(path + '/' + 'station_final.csv', encoding='ansi', index=False)
```
