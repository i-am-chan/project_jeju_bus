```python
import pandas as pd
```

```python
prefix_path = 'D:/jeju_bus_data_no_leakage/bis/'
file_name = 'BIS_190531.csv'

path = prefix_path + file_name
```

```python
df = pd.read_csv(path, 
                 usecols = ['EVENT_DATE', 'EVENT_CD', 'VH_ID', 'ROUTE_ID', 
                            'BUS_RUN_CNT', 'ROUTE_RUN_CNT', 'STATION_ORD', 'RUN_DIST'])
```

```python
df = df[(df['EVENT_CD'] == 18) | (df['EVENT_CD'] == 17)]
```

```python
grouped_1 = df.groupby(['ROUTE_ID', 'ROUTE_RUN_CNT', 'VH_ID', 'BUS_RUN_CNT'])
```

```python
# import IPython.display

cnt = 0
g = []
for name, group in grouped_1:
    group = group.sort_values(by=['EVENT_DATE'])
    if group.iloc[-1]['EVENT_CD'] == 18:
        #display(group.iloc[[-1]])
        print(str(group.iloc[len(group) - 1]).replace('\n', ' '))
        g.append(group)
        print(name)
        cnt+=1
#     print(group.iloc[len(group) - 1]['EVENT_CD'])
#     cnt += 1
print(cnt)
```
